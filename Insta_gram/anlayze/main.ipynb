```python
from sentiment_analyze import KoreanSentimentAnalyzer
```

```python
csv_path = "/home/kang/PythonProject/Project 1./Text_Project/Insta_gram/DATA/with_nouns_adjs.csv"
json_path = "/home/kang/PythonProject/Project 1./Text_Project/News/data/SentiWord_info.json"

analyzer = KoreanSentimentAnalyzer(csv_path, json_path)
```

```python
# 각 후보별 평균 감성 점수 추출

score = analyzer.analyze_keyword_sentiment("이재명")

print(f"이재명의 평균 감성 점수: {score}")
```

```python
score = analyzer.analyze_keyword_sentiment("김문수")

print(f"김문수의 평균 감성 점수: {score}")
```

```python
score = analyzer.analyze_keyword_sentiment("이준석")

print(f"이준석의 평균 감성 점수: {score}")
```

```python
keywords = ["이재명", "김문수", "이준석"]

for keyword in keywords:
    score = analyzer.analyze_keyword_sentiment(keyword)
    print(f"{keyword}의 평균 감성 점수: {score:.3f}")
```
